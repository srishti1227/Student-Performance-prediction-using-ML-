```python
#3.7 Adding the columns of Total and Average 
df['Total Score'] = df['math_score'] + df['writing_score'] + df['reading_score']
df['Average Score '] = df['Total Score']/3
print(df)
```

# 4. Exploring Dataset(Visualization)

#in this section we will be using visualization library in order to visualize certian data wrt to other factors


```python
fig, axs = plt.subplots(1, 2, figsize = (15,7))
plt.subplot(121)
sns.histplot(data = df, x = df['Average Score '], kde = True, color = 'g', bins= 30)
plt.subplot(122)
sns.histplot(data = df, x = df['Average Score '], kde=True, bins = 30, hue = 'gender', color = 'g')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize = (15,7))
plt.subplot(121)
sns.histplot(data = df, x = df['math_score'], kde = True, color = 'g', bins= 30)
plt.subplot(122)
sns.histplot(data = df, x = df['math_score'], kde=True, bins = 30, hue = 'gender', color = 'g')
plt.show()
```

Insights: 
1. Female perform better than man 
2. in avg score in maths female has higher maths score,but boys with 100 marks are more than females 

```python
#let's now see, what effect does lunch before the test have, both on males and females
fig,axs = plt.subplots(1,3, figsize = (25,7))
plt.subplot(141)
sns.histplot(data = df, x = 'Average Score ', kde = True, color = 'g', bins = 30, hue = 'lunch')
plt.subplot(142)
sns.histplot(data = df[df.gender == 'female'], x = 'Average Score ', kde = True, bins = 30, hue = 'lunch')
plt.subplot(143)
sns.histplot(data = df[df.gender == 'male'], x = 'Average Score ', hue = 'lunch', kde = True, bins = 30)
plt.show()  
```

Insights from the data
1. So having lunch before the exam helps score better for both make and female
